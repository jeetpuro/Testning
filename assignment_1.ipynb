### 2.1 Statement coverage
Compute the statement coverage of the program using [`coverage.py`](https://coverage.readthedocs.io/en/latest/index.html). (**2 points**)

```python
!coverage run estimate_pi.py
# !coverage run test_estimate_pi.py

!coverage report -m
# 18-19, 22->exit, 29, 34
```

How can we interprete the results?

    Interpretation
    We can see how much of the main program the test really tests. Here we see what rows are being run when the functions are being called and the rows that are being missed from the file. Coverage shows us the percentage of rows that are being run by the file.
